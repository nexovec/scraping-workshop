```python
links = []
for element in elements:
    link_element = element.find("a") # doplň tag pro odkazy
    link = link_element.get("href") ## doplň atribut pro odkazy (URL adresu)
    links.append(link)
```

```python
urls = []
for link in links:
    url = domain + link
    urls.append(url)
    
urls
```

```python
kocky = []

#navštívíme každý připravený odkaz a stáhneme si informace o kočkách
for url in urls:
    response = requests.get(url)

    if response.status_code == 200:
        html_content = response.text

    else:
        print(f"Chyba při načítání stránky. HTTP kód: {response.status_code}")
        
    soup = bs(response.text)
    
    #vytvoříme si slovník pro uchování informací o kočce
    charakteristika = {}
        
    #najdeme jméno kočky
    jmeno = soup.find("title").text
    charakteristika["title"] = jmeno
    
    #uložíme si všechny informace, které ke kočce jsou k dispozici
    elements = soup.find_all('div', class_='tags')
    
    for element in elements:
        if len(element.text.strip().split("\t")) == 1:
            pass
        else:
            key, value = element.text.replace("\n", " ").strip().split("\t")
            key = " ".join(key.split())
            value = " ".join(value.split())
            charakteristika[key] = value


    kocky.append(charakteristika)

    sleep(5)
```
